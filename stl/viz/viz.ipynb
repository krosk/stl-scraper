```python
import plotly.graph_objects as go
import numpy as np

# Standardize the 'total_price' column
total_price_mean = with_reviews['total_price'].mean()
total_price_std = with_reviews['total_price'].std()
with_reviews['standardized_price'] = (with_reviews['total_price'] - total_price_mean) / total_price_std
# absolute value
with_reviews['standardized_price'] = with_reviews['standardized_price'].abs()


# Replace NaN values in 'avg_rating' with a default opacity value
default_opacity = 0.5  # Choose a default opacity value (between 0 and 1)
with_reviews['avg_rating'] = with_reviews['avg_rating'].fillna(default_opacity)

# Create a hovertemplate string
hovertemplate = (
    "<b>Name</b>: %{text}<br>"
    "<b>Price</b>: %{marker.color:.2f}<br>"
    "<b>Person capacity</b>: %{customdata[0]}<br>"
    "<b>Ratings</b>:<br>"
    "- Count: %{customdata[1]}<br>"
    "- Checkin: %{customdata[2]:.2f}<br>"
    "- Cleanliness: %{customdata[3]:.2f}<br>"
    "- Communication: %{customdata[4]:.2f}<br>"
    "- Location: %{customdata[5]:.2f}<br>"
    "- Value: %{customdata[6]:.2f}<br>"
    "<b>No. of photos</b>: %{customdata[7]}<br>"
    "<extra></extra>"
)

# Create a scattermapbox trace
trace = go.Scattermapbox(
    lat=with_reviews['latitude'],  # Latitude values for each data point
    lon=with_reviews['longitude'],  # Longitude values for each data point
    mode='markers',
    marker=dict(
        size=14 + 2.5 * with_reviews['standardized_review_count'],  # Adjust the size range based on standardized 'total_price' values
        color=with_reviews['total_price'],  # 'total_price' column values for each data point
        colorscale='Viridis',  # You can choose any color scale you prefer
        colorbar=dict(title='Total Price'),  # Add a colorbar with a title
        opacity=with_reviews['avg_rating'] / with_reviews['avg_rating'].max(),  # Set opacity based on 'avg_rating' values
        cmin=with_reviews['total_price'].min(),  # Set the minimum value for color scale normalization
        cmax=with_reviews['total_price'].max(),  # Set the maximum value for color scale normalization
        colorbar_ticksuffix=' zł',  # Add a prefix for colorbar tick values
        colorbar_tickformat='.2f',  # Set the format for colorbar tick values
    ),
    text=with_reviews['name'],  # Text labels for each data point
    customdata=np.array([
        with_reviews['person_capacity'],
        with_reviews['review_count'],
        with_reviews['rating_checkin'],
        with_reviews['rating_cleanliness'],
        with_reviews['rating_communication'],
        with_reviews['rating_location'],
        with_reviews['rating_value'],
        with_reviews['photo_count'],
    ]).T,
    hovertemplate=hovertemplate,
)

# Create the layout for the map
layout = go.Layout(
    mapbox=dict(
        style='open-street-map',  # Choose the map style (e.g., 'open-street-map', 'carto-positron', etc.)
        center=dict(
            lat=with_reviews['latitude'].mean(),  # Use the mean latitude value for the map center
            lon=with_reviews['longitude'].mean(),  # Use the mean longitude value for the map center
        ),
        zoom=13,  # Adjust the zoom level based on your data and preferences
    ),
    
)


# Create the figure and add the trace to it
fig = go.Figure(data=[trace], layout=layout)

# set size to fit 2/3 column width -> 
fig.update_layout(
    autosize=False,
    width=800,
    height=800,
    margin=dict(
        # no margin
        l=0,
        r=0,
        b=0,
        t=0,
        pad=0
    ),
    paper_bgcolor="Ivory", # where to find plotly color names: https://www.w3schools.com/colors/colors_names.asp
)


# Show the figure
fig.show()
```

```python
# add callbacks to the map - filter data

import plotly.graph_objects as go
# Output is not defined yet
from dash.dependencies import Input, Output

@app.callback(
    Output("map-graph", "figure"),
    [
        Input("avg_rating-slider", "value"),
        Input("total_price-slider", "value"),
        Input("person_capacity-slider", "value"),
        Input("rating_accuracy-slider", "value"),
        Input("rating_checkin-slider", "value"),
        Input("rating_cleanliness-slider", "value"),
        Input("rating_communication-slider", "value"),
        Input("rating_location-slider", "value"),
        Input("rating_value-slider", "value"),
        Input("review_count-slider", "value"),
    ],
)
def update_graph(
    avg_rating_value,
    total_price_value,
    person_capacity_value,
    rating_accuracy_value,
    rating_checkin_value,
    rating_cleanliness_value,
    rating_communication_value,
    rating_location_value,
    rating_value_value,
    review_count_value,
):
    # Filter the data
    filtered_df = with_reviews[
        (with_reviews["avg_rating"] >= avg_rating_value[0])
        & (with_reviews["avg_rating"] <= avg_rating_value[1])
        & (with_reviews["total_price"] >= total_price_value[0])
        & (with_reviews["total_price"] <= total_price_value[1])
        & (with_reviews["person_capacity"] >= person_capacity_value[0])
        & (with_reviews["person_capacity"] <= person_capacity_value[1])
        & (with_reviews["rating_accuracy"] >= rating_accuracy_value[0])
        & (with_reviews["rating_accuracy"] <= rating_accuracy_value[1])
        & (with_reviews["rating_checkin"] >= rating_checkin_value[0])
        & (with_reviews["rating_checkin"] <= rating_checkin_value[1])
        & (with_reviews["rating_cleanliness"] >= rating_cleanliness_value[0])
        & (with_reviews["rating_cleanliness"] <= rating_cleanliness_value[1])
        & (with_reviews["rating_communication"] >= rating_communication_value[0])
        & (with_reviews["rating_communication"] <= rating_communication_value[1])
        & (with_reviews["rating_location"] >= rating_location_value[0])
        & (with_reviews["rating_location"] <= rating_location_value[1])
        & (with_reviews["rating_value"] >= rating_value_value[0])
        & (with_reviews["rating_value"] <= rating_value_value[1])
        & (with_reviews["review_count"] >= review_count_value[0])
        & (with_reviews["review_count"] <= review_count_value[1])
    ]

    # Create a hovertemplate string
    hovertemplate = (
        "<b>Name</b>: %{text}<br>"
        "<b>Price</b>: %{marker.color:.2f}<br>"
        "<b>Person capacity</b>: %{customdata[0]}<br>"
        "<b>Ratings</b>:<br>"
        "- Count: %{customdata[1]}<br>"
        "- Checkin: %{customdata[2]:.2f}<br>"
        "- Cleanliness: %{customdata[3]:.2f}<br>"
        "- Communication: %{customdata[4]:.2f}<br>"
        "- Location: %{customdata[5]:.2f}<br>"
        "- Value: %{customdata[6]:.2f}<br>"
        "<b>No. of photos</b>: %{customdata[7]}<br>"
        "<extra></extra>"
    )

    # Create a scattermapbox trace
    trace = go.Scattermapbox(
        lat=filtered_df["latitude"],  # Latitude values for each data point
        lon=filtered_df["longitude"],  # Longitude values for each data point
        mode="markers",
        marker=dict(
            size=14 + 2.5 * filtered_df["standardized_review_count"],  # Adjust the size range based on standardized 'total_price' values
            color=filtered_df["total_price"],  # 'total_price' column values for each data point
            colorscale="Viridis",  # You can choose any color scale you prefer
            colorbar=dict(title="Total Price"),  # Add a colorbar with a title
            opacity=filtered_df["avg_rating"] / filtered_df["avg_rating"].max(),  # Set opacity based on 'avg_rating' values
            cmin=filtered_df["total_price"].min(),  # Set the minimum value for color scale normalization
            cmax=filtered_df["total_price"].max(),  # Set the maximum value for color scale normalization
            colorbar_ticksuffix=" zł",  # Add a prefix for colorbar tick values
            colorbar_tickformat=".2f",  # Set the format for colorbar tick values
        ),
        text=filtered_df["name"],  # Text labels for each data point
        customdata=np.array([
            filtered_df["person_capacity"],
            filtered_df["review_count"],
            filtered_df["rating_checkin"],
            filtered_df["rating_cleanliness"],
            filtered_df["rating_communication"],
            filtered_df["rating_location"],
            filtered_df["rating_value"],
            filtered_df["photo_count"],
        ]).T,

        hovertemplate=hovertemplate,
    )

    # # Create the layout for the map
    # layout = go.Layout(
    #     # adjust size of themap to fit the card
    #     # height='100%',
    #     mapbox=dict(
    #         style="carto-positron",  # Choose the map style (e.g., 'open-street-map', 'carto-positron', etc.)
    #         center=dict(
    #             lat=filtered_df["latitude"].mean(),  # Use the mean latitude value for the map center
    #             lon=filtered_df["longitude"].mean(),  # Use the mean longitude value for the map center
    #         ),
    #         zoom=13,  # Adjust the zoom level based on your data and preferences
    #         # fit map to card

    #     ),
    #     title="Mapbox Visualization",
    # )

    

    # Create the figure and add the trace to it
    fig = go.Figure(data=[trace], layout=layout)

    # Return the figure
    return fig

# show app
app.run_server(use_reloader=False)
```
